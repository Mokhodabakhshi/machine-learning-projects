```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_excel("CSM_dataset.xlsx",na_values="NaN").set_axis(list(range(14)),axis=1)
data
```

```python
# discrete attribute and continous attribute
dist_attr = [0]
cont_attr = [x for x in range(1,14)]
```

```python
a = list(data.isna().sum())
nan_value = [i for i,_ in enumerate(a) if _ != 0]
```

```python
# missing values
for i in nan_value:
    _mean = data[i].mean()
    data[i] = data[i].replace(math.nan, _mean)
```

```python
data.isna().sum()
```

```python
attr = [x for x in cont_attr if x != 4]
```

```python
# min_max_scaler = MinMaxScaler()
# x = data[attr].values
# x_scaled = min_max_scaler.fit_transform(x)
# normalized = pd.DataFrame(x_scaled, index=data.index, columns=attr)
# print(normalized.shape)
# print(x.shape)
# data[attr] = normalized
```

```python
std = StandardScaler()
x = data[attr].values
x_scaled = std.fit_transform(x)
normalized = pd.DataFrame(x_scaled, columns=attr, index=data.index)
data[attr] = normalized
```

```python
X = data[attr]
Y = data.pop(4)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2)
```

```python
lr_sgd = SGDRegressor()
lr_sgd.fit(x_train, y_train)
score = lr_sgd.score(x_test,y_test)
y_pred = lr_sgd.predict(x_test)
print(score)
```

```python
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
```
